```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```
X = df.drop(columns=['Target'])
y = df['Target']
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=10)
```

```
print(f"The shape and percentages of X_train is: {X_train.shape} and {len(100*X_train)/len(df)}%")
print(f"The shape and percentages of X_train is: {X_test.shape} and {len(100*X_test)/len(X)}%")
print(f"The shape and percentages of X_train is: {y_train.shape} and {len(100*y_train)/len(X)}%")
print(f"The shape and percentages of X_train is: {y_test.shape} and {len(100*y_test)/len(X)}%")
```

```
clf = RandomForestClassifier(n_estimators=100, random_state=10, max_features='sqrt', n_jobs=-1, verbose=1)
```

```
clf
```

# Column Transformers, Encoders, scalers and pipeline

```
cat_cols = [col for col in X.select_dtypes(include='object').columns]
num_cols = [col for col in X.select_dtypes(exclude='object').columns]
print("Categorical==>", cat_cols, "\n",'Numerical==>', num_cols)
```

```
ohe = OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```
ct = ColumnTransformer([
                   ('step1', MinMaxScaler(), num_cols),
                   ('step2', ohe, cat_cols)
                   ], remainder = 'passthrough')
ct
```

```
#pipeline everything
my_pipe = Pipeline([
                    ('coltran', ct),
                    ('clf', clf)
])
my_pipe
```

```
my_pipe.fit(X_train, y_train)
```

```
y_pred = my_pipe.predict(X_test)
y_pred
```

```
from sklearn.metrics import roc_auc_score, confusion_matrix
```

```
roc_auc_score(y_test, y_pred)
```

```
confusion_matrix(y_test, y_pred)
```
