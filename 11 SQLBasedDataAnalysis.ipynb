### Data Analysis
* Create age and score histogram
* Re-order students
* Re-order students by faculty
* Average age by faculty
* Average score by faculty

```python
# Method 1: age Histogram
pd.read_sql_query("""With t as ( SELECT CAST((age/10)*10 AS INTEGER) as bin, age FROM students) 
                     SELECT bin, count(*)FROM t Group BY bin """, conn) 
```

```python
# Method 2: age Histogram: in this case the count includes the upper limit
pd.read_sql_query("""
With t as (SELECT CASE WHEN age BETWEEN 15 AND 20 THEN '10-20' 
            WHEN age BETWEEN 20 AND 25 THEN '20-30'
            WHEN age BETWEEN 25 AND 30 THEN '30-40' 
            WHEN age BETWEEN 30 AND 35 THEN '40-50'
            WHEN age BETWEEN 35 AND 40 THEN '50-60'
            ELSE '60+' END as bin, age FROM students)
SELECT bin, COUNT(*) FROM t GROUP BY bin
""", conn)
```

```python
#Dense Rank
pd.read_sql_query("SELECT DENSE_RANK() OVER(ORDER BY faculty) as dranknum, * FROM students", conn)
```
